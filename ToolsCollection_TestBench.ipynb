```python
import ROOT as r
from ToolsCollection import fft,fft2
```

```python
func = r.TF1("func","1+0.5*cos(6*x+1.5)",0,20);
h1 = r.TH1D("h1","h1;t;y",100,0,20);
h1.SetLineColor(2);
h1.FillRandom("func",10000);
hfft = r.fft(h1,"new_fft","title for fft",0.01);

c = r.TCanvas("c","c",1200,600);
c.Divide(2,1);
c.cd(1).SetGrid();
h1.Draw("hist");
c.cd(2).SetGrid();
hfft.Draw("hist");
hfft.GetYaxis().SetMaxDigits(2)
c.Draw();
```

```python
c = r.TCanvas("c","c",1800,600);
c.Divide(3,1);

c.cd(1).SetGrid();
hfft1 = r.fft2(h1,"new_fft1","title for fft",0.01,"RE");
hfft1.GetYaxis().SetMaxDigits(2)
hfft1.Draw("hist");

c.cd(2).SetGrid();
hfft2 = r.fft2(h1,"new_fft2","title for fft",0.01,"IM");
hfft2.GetYaxis().SetMaxDigits(2)
hfft2.Draw("hist");

c.cd(3).SetGrid();
hfft3 = r.fft2(h1,"new_fft3","title for fft",0.01,"PH");
hfft3.GetYaxis().SetMaxDigits(2)
hfft3.Draw("hist");

c.Draw();
```
